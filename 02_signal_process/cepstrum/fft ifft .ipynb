```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs =102400
total_duration = 0.01

signal_duration = 0.001

x = np.ones(int(signal_duration*fs))

length = (int(total_duration*fs)-len(x))//2

signal = np.concatenate((np.zeros(length), x, np.zeros(length)))

time = np.arange(len(signal))/fs
fig, ax = plt.subplots()
ax.plot(time, signal)
```

```python
X = (np.fft.fft(signal))

ix = np.abs(np.fft.ifft(X))
```

```python
fig, ax = plt.subplots()
ax.plot(time, ix)
```
