```python
import torch 
import numpy as np 
from torch.autograd import Variable
```

```python
numpy = np.array(([.1,2], [3,4]))
tensor = torch.from_numpy(numpy)
t2n = tensor.numpy()

numpy[0][0] = 100
tensor[1][0] = -100
t2n[1][1] = -100 

print("numpy: \n", numpy)
print("tensor: \n", tensor)
print("tensor to numpy: \n", t2n)
```

the trasformation between tensor and numpy, but it seems like they are related together 

```python
from torchviz import make_dot
variable = Variable(tensor, requires_grad = True)
v_out = (variable*variable).mean()
make_dot(v_out)
```

```python
v_out.backward()
variable.grad
```
