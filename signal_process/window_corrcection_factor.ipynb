```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.fftpack import fft, fftshift, fftfreq
from scipy.signal.windows import hann, hamming
```

```python
fs = 10
t = np.arange(0, 2, 1/fs)
f0 = 1
x = np.cos(np.pi*2*t*f0)
N = len(x)
w = hamming(N)

xw = x*w

xk = fftshift(fft(x))
xwk = fftshift(fft(xw))

freq = fftshift(fftfreq(N, 1/fs))
plt.figure(figsize=(6.4*2, 4.8))
plt.bar(freq, abs(xk), label="origin signal", width=0.05)

plt.bar(freq, abs(xwk), label="windowed signal", width=0.05)
plt.legend()
plt.show()
```

the only way to make sure how the window scaled is to test it out

```python
aef = abs(xk).max()/abs(xwk).max()
```

```python
print("amplitude correction factor: {}".format(aef))
```

```python
def zero_padding(x, length):
    if(len(x)>=length):
        print("Error: desired lengh is smaller than origin length")
        return x
    elif(len(x) == length):
        return x
    else:
        z = np.zeros(length-len(x))
        return np.concatenate((x, z), axis=0)
    
```
