#### matplotlib is object hierarchy

1. figure is the outer container
2. with multiple Axes

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng = np.arange(50)
rnd = np.random.randint(0, 10, size=(3, rng.size))
yrs = 1950 + rng

fig, ax = plt.subplots(figsize=(5, 3))
ax.stackplot(yrs, rng + rnd, labels=('Eastasia', 'Eurasia', 'Oceania'))
ax.set_title('Combined debt growth over time')
ax.legend(loc='upper left')
ax.set_ylabel('Total debt')
ax.set_xlim(xmin=yrs[0], xmax=yrs[-1])
fig.tight_layout()
```

## Pseudo random VS true random

random number that is generated by code called pseudo random
code that generate number by getting the random data called true random

### randint()
return int or int of array with the discrete uniform distribution integer that is within the range [low, high)
### rand()
return float or float of array with uniform distribution over [0, 1)
### randn()
return float or float of array from standard normal distribution

```python
# print(plt.rcParams) # to examine all values
 
print(plt.rcParams.get('figure.figsize'))
```

```python
x = np.random.randint(low=1, high=11, size=50)
y = x + np.random.randint(1, 5, size=x.size)
data = np.stack((x,y), axis=1)

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2,
                               figsize=(8, 4))

ax1.scatter(x=x, y=y, marker='o', c='r', edgecolor='b')
ax1.set_title('Scatter: $x$ versus $y$')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')

ax2.hist(data, bins=np.arange(data.min(), data.max()+1),
         label=('x', 'y'))
ax2.legend(loc=(0.65, 0.8))
ax2.set_title('Frequencies of $x$ and $y$')
ax2.yaxis.tick_right()
```
