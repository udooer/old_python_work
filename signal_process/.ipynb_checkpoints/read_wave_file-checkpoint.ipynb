```python
# another useable function from scipy, it can get the sampling freq and numpy array data 
from scipy.io.wavfile import read
```

```python
fs, data = read("dolphinsound_20_sec.wav")
data.shape
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
ch1 = data[:,0]
tstart = 15*fs
te = 15.1*fs
tend = tstart + int(math.pow(2, math.ceil(math.log2(te-tstart))))

ch1 = ch1[tstart:tend]
# pad = np.zeros(int(math.pow(2, next2)-N))
# ch1 = np.concatenate((ch1, pad))
plt.figure(figsize=(6.4*3, 4.8))
plt.plot(ch1)
plt.xlabel("t()")
plt.ylabel("amplitude")
plt.title("dolphin sound from {}s to {}s".format(tstart/fs, tend/fs))
plt.show()
len(ch1)
```

```python
N = len(ch1)
ch1k = np.fft.fft(ch1)
absch1 = np.abs(ch1k)/N*2
f = np.arange(0, fs, fs/N)
plt.bar(f[:N//2+1], absch1[:N//2+1])
plt.xlabel("Hz")
plt.ylabel("Amplitude")
plt.show()
```
