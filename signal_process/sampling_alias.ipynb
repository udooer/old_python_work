```python
import numpy as np
# Number of sample points
N = 500
# sample spacing
T = 1.0 / (590*10)
x_low = np.linspace(0.0, N*T, N, endpoint=False)
y_low = np.sin(50.0 * 2.0*np.pi*x_low) + 0.5*np.sin(80.0 * 2.0*np.pi*x_low) + np.sin(590 * 2.0*np.pi*x_low)

x_high = np.linspace(0.0, N*T, N*100, endpoint=False)
y_high = np.sin(50.0 * 2.0*np.pi*x_high) + 0.5*np.sin(80.0 * 2.0*np.pi*x_high) + np.sin(590 * 2.0*np.pi*x_high)
import matplotlib.pyplot as plt
plt.figure(figsize=(6.4*3, 4.8))
plt.plot(x_low[:50], y_low[:50])
plt.plot(x_high[:5000], y_high[:5000])

plt.show()
```

```python
import numpy as np
# Number of sample points
N = 500
# sample spacing
T = 1.0 / (800)
x_sample = np.linspace(0.0, N*T, N, endpoint=False)
y_sample = np.sin(590 * 2.0*np.pi*x_sample)

x_true = np.linspace(0.0, N*T, N*100, endpoint=False)
y_true = np.sin(590 * 2.0*np.pi*x_true)

x_think = np.linspace(0.0, N*T, N*100, endpoint=False)
y_think = np.sin(-210 * 2.0*np.pi*x_think)
import matplotlib.pyplot as plt
plt.figure(figsize=(6.4*3, 4.8))
plt.plot(x_true[:2000], y_true[:2000], 'y-')
plt.plot(x_think[:2000], y_think[:2000], 'r-')
plt.bar(x_sample[:20], y_sample[:20], width=0.00005, color='k')
plt.grid()
plt.legend(['590(Hz) origin signal', 'the sampling data which computer think is(210Hz)', 'sampling data bar'], loc='lower center')

plt.show()
```
