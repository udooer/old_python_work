```python
import numpy as np
from scipy.fftpack import fft
# Number of sample points
N = 800
# sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N, endpoint=False)
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
yf = fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2, endpoint=False)
import matplotlib.pyplot as plt
plt.figure()
plt.bar(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.show()
```

```python
print("the frequency is: {}".format(xf[yf.argmax()]), "\nthe freq is {} * {}".format(yf.argmax(), xf[1]-xf[0]))
# print("the delta of the freq: {}".format(xf[1]-xf[0]))
```

```python
from scipy.fftpack import fft
import numpy as np
# Number of sample points
N = 500
# sample spacing
T = 1.0 / 1600 # 800, 1000, 1500, 1600, 2000, 4000
x = np.linspace(0.0, N*T, N, endpoint=False)
x_high = np.linspace(0.0, N*T, N*1000, endpoint=False)

y_high = np.sin(50.0 * 2.0*np.pi*x_high) + 0.5*np.sin(80.0 * 2.0*np.pi*x_high) + np.sin(590 * 2.0*np.pi*x_high) + np.sin(750 * 2.0*np.pi*x_high)
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x) + np.sin(590 * 2.0*np.pi*x) + np.sin(750 * 2.0*np.pi*x)
yf = fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2, endpoint=False)
import matplotlib.pyplot as plt
plt.figure(figsize=(6.4*2, 4.8*3))
plt.subplot(3, 1, 1)
plt.plot(x[:50], y[:50], 'r')
plt.plot(x_high[:50000], y_high[:50000], 'y')
plt.subplot(3, 1 ,2)
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.subplot(3, 1, 3)
plt.bar(xf, 2.0/N * np.abs(yf[0:N//2]), width=1.5) # width will effect the plot
plt.show()
```

```python
from scipy.fftpack import fft
from scipy.signal import blackman

# Number of sample points
N = 500
# sample spacing
T = 1.0 / 1600
x = np.linspace(0.0, N*T, N, endpoint=False)

x_high = np.linspace(0.0, N*T, N*1000, endpoint=False)

y_high = np.sin(50.0 * 2.0*np.pi*x_high) + 0.5*np.sin(80.0 * 2.0*np.pi*x_high) + np.sin(590 * 2.0*np.pi*x_high) + np.sin(750 * 2.0*np.pi*x_high)
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x) + np.sin(590 * 2.0*np.pi*x) + np.sin(750 * 2.0*np.pi*x)

yw = y*blackman(N)

yf = fft(y)
ywf = fft(yw)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2, endpoint=False)
import matplotlib.pyplot as plt
plt.figure(figsize=(6.4*2, 4.8*4))
plt.subplot(4, 1, 1)
plt.plot(x[:50], y[:50], 'r')
plt.plot(x_high[:50000], y_high[:50000], 'y')
plt.subplot(4, 1, 2)
plt.plot(x, yw)
plt.subplot(4, 1 ,3)
plt.semilogy(xf, 2.0/N * np.abs(ywf[0:N//2]))
plt.semilogy(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.subplot(4, 1, 4)
plt.bar(xf, 2.0/N * np.abs(ywf[0:N//2]), width=1.5)
plt.show()
```
